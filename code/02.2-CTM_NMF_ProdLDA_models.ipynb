```python
import gensim
import pandas as pd
df = pd.read_csv('../Topic_modeling/code_data/data/dataset_media.csv')

df = df.drop_duplicates(subset='tweet_id')
print('-- Dataframe shape: {}'.format(df.shape))
df = df.groupby('topic').apply(lambda x: x.sample( len(x) if len(x) < 10000 else 10000,random_state=42)).reset_index(drop=True)
df = df.reset_index(drop=True)


df['processed_text2'] = df['processed_text'].apply(lambda x: set(x.split()))

docs = df.processed_text2.to_numpy()

# create dictionary of all words in all documents
dictionary = gensim.corpora.Dictionary(docs)

# create variable containing length of dictionary/vocab
vocab_length = len(dictionary)
if input('Do you want save vocabs on vocab.txt?[y/N]')=='y':
    with open("../data/media/vocabulary.txt", "w",encoding='utf8') as f:
        for i in range(vocab_length):
            f.write(dictionary[i])
            f.write('\n')
# create BOW dictionary 
bow_corpus = [dictionary.doc2bow(doc) for doc in docs]
df['corpus']=bow_corpus

from sklearn.model_selection import train_test_split
index_train, index_test= train_test_split(list(df.index.values), test_size=0.20, random_state=42)
index_train,index_valid = train_test_split(index_train, test_size=0.20, random_state=42)

df.loc[index_train , 'div'] = 'train'
df.loc[index_valid , 'div'] = 'val'
df.loc[index_test , 'div'] = 'test'

d2 = df[['processed_text','div','topic']]

d2.to_csv("../data/media/corpus.tsv", index=False, sep='\t')
print('[ OK ] Dataset created.')
```

```python
import numpy as np
from collections import Counter
from sklearn import metrics
from gensim.models import CoherenceModel

def q_metrics(y_true, y_pred,my_model=None):
    contigency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    purity = np.sum(np.amax(contigency_matrix, axis=0)) / np.sum(contigency_matrix)

    print('purity_score:',purity)
    print('NMI:',metrics.normalized_mutual_info_score(y_true, y_pred))
    if my_model!=None:
        cm = CoherenceModel(model=my_model, corpus=bow_corpus, dictionary=dictionary, coherence='u_mass')
        print('Coherence:',cm.get_coherence())
```

```python
from octis.models.ProdLDA import ProdLDA
from octis.models.LDA import LDA
from octis.models.NMF import NMF
from octis.models.CTM import CTM

from octis.dataset.dataset import Dataset
from octis.evaluation_metrics.diversity_metrics import TopicDiversity
from octis.evaluation_metrics.coherence_metrics import Coherence

dataset = Dataset()
dataset.load_custom_dataset_from_folder("../data/media/")
```

```python
model = ProdLDA(num_topics=11,batch_size=128, num_epochs=13,num_neurons=100,num_layers=2,lr=0.002)

# model = LDA(iterations=20 , num_topics=11, 
#            chunksize=2000, random_state=None, gamma_threshold=0.001,
#            passes=1, update_every=1,
#            alpha="symmetric")

# model =  CTM(batch_size=64,num_topics=11, num_epochs=2,
#              inference_type='zeroshot',bert_model="all-MiniLM-L6-v2")


# model = NMF(num_topics=11, chunksize=2000, passes=1, kappa=1.0,
#            minimum_probability=0.01, w_max_iter=200,
#            w_stop_condition=0.0001, h_max_iter=50, h_stop_condition=0.001,
#            eval_every=10, normalize=True,
#            random_state=None)

output = model.train_model(dataset)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('../data/media/corpus.tsv',sep='\t')


resres = output['test-topic-document-matrix'].T
pred = []
for i in range(len(resres)):
    pred.append(np.argmax(resres[i]))

y_true = df[df['div']=='test']['topic']
y_pred = pred
q_metrics(y_true, y_pred)

npmi = Coherence(texts=dataset.get_corpus(), topk=10, measure='c_v')

topic_diversity = TopicDiversity(topk=10)

topic_diversity_score = topic_diversity.score(output)
print("Topic diversity: "+str(topic_diversity_score))

npmi_score = npmi.score(output)
print("Coherence: "+str(npmi_score))



#ProdLDA
"""purity_score: 0.35801912773521605
NMI: 0.19716537375142443
Topic diversity: 0.9272727272727272
Coherence: 0.5578935064200444"""

#LDA
"""purity_score: 0.31650893796004204
NMI: 0.17364713425105782
Topic diversity: 0.8272727272727273
Coherence: 0.4831534719385152"""

#NMF
"""purity_score: 0.2968304040859246
NMI: 0.14278633732286053
Topic diversity: 0.7
Coherence: 0.5330212153474894"""

# CTM
"""
purity_score: 0.4337289069150268
NMI: 0.24710445841683407
Topic diversity: 0.9545454545454546
Coherence: 0.6111417234433117
"""
```
